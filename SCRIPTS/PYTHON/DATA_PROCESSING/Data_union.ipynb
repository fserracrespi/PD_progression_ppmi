# union without datscan

```python
# Merge
data = subjects_df.merge(non_motor_df, on=['PATNO','EVENT_ID'], how="outer")
data = data.merge(updrs_df, on=['PATNO','EVENT_ID'], how="outer")

print('NaN values in data dataset (antes):', data.isna().sum().sum())
print('Shape (antes):', data.shape, '\nUnique Patnos:', data['PATNO'].nunique())

# =========================================================
# IMPUTACIÓN (todas categóricas):
# 1) Reemplazar 101 -> NaN (en todas las columnas excepto keys)
# 2) Eliminar filas con NHY NaN
# 3) Eliminar filas con >30% faltantes (sobre todas las cols excepto keys)
# 4) Imputar NaNs restantes con la MODA por columna
#    (si una columna queda completamente NaN, se deja igual)
# =========================================================

import numpy as np

keys = ['PATNO', 'EVENT_ID']
cols = [c for c in data.columns if c not in keys]

# 1) 101 -> NaN (si 101 es tu código de missing)
data[cols] = data[cols].replace(101, np.nan)

# 2) Eliminar filas donde NHY sea NaN (asumiendo que NHY existe en data)
data = data.loc[data['NHY'].notna()].copy()
print('Shape (después de eliminar NHY NaN):', data.shape)

# 3) Eliminar filas con >30% faltantes (solo en cols)
nan_ratio = data[cols].isna().mean(axis=1)   # <-- CORRECCIÓN AQUÍ
data = data.loc[nan_ratio <= 0.30].copy()
print('Shape (después de eliminar >30% NaNs):', data.shape)

# 4) Imputar con moda por columna
for c in cols:
    moda = data[c].mode(dropna=True)
    if not moda.empty:
        data[c] = data[c].fillna(moda.iloc[0])

print('NaN values in data dataset (después):', data.isna().sum().sum())
print('Shape (final):', data.shape, '\nUnique Patnos:', data['PATNO'].nunique())

data.head()
```

```python
dict_visits={'type1':['BL','V04','V06','V08','V10'],
             'type2':['BL','V04','V06','V08'],
             'type3':['BL','V04','V06'],
             'type4':['BL','V04'],
             'type_special':['V04','V06','V08','V10']}

counter_type={'type1':{'counter':0,'patnos':[],'type_users':{"Parkinson's Disease":0,'Prodromal':0}},
              'type2':{'counter':0,'patnos':[],'type_users':{"Parkinson's Disease":0,'Prodromal':0}},
              'type3':{'counter':0,'patnos':[],'type_users':{"Parkinson's Disease":0,'Prodromal':0}},
              'type4':{'counter':0,'patnos':[],'type_users':{"Parkinson's Disease":0,'Prodromal':0}},
              'type_special':{'counter':0,'patnos':[],'type_users':{"Parkinson's Disease":0,'Prodromal':0}},
              'type_other':{'counter':0,'patnos':[],'type_users':{"Parkinson's Disease":0,'Prodromal':0}}}

for patno in data['PATNO'].unique():

    patno_data=data[data['PATNO']==patno]
    visits=patno_data['EVENT_ID'].unique().tolist()


    for key, value in dict_visits.items():
        if all(visit in visits for visit in value):
            counter_type[key]['counter']+=1
            counter_type[key]['patnos'].append(patno)
                # Count the type of user (PD or Prodromal)
            user_type=patno_data['COHORT_DEFINITION'].iloc[0]
            counter_type[key]['type_users'][user_type]+=1
    else:
        counter_type['type_other']['counter']+=1
        counter_type['type_other']['patnos'].append(patno)

for key in counter_type:
    print(f"{key}: {counter_type[key]['counter']} patients", f"({counter_type[key]['type_users']})")
```
