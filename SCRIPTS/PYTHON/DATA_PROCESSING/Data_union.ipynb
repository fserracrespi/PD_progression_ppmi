```python
import pandas as pd
from pathlib import Path

current_dir = Path.cwd()
project_root = current_dir.parents[2]

datscan_path = project_root / "DATA_PPMI/Results/DatScan_SBR.csv"
updrs_path = project_root / "DATA_PPMI/Results/MDS_UPDR.csv"
non_motor_path = project_root / "DATA_PPMI/Results/NON_MOTOR_DATA.csv"
subjects_path = project_root / "DATA_PPMI/Results/SUBJECT_CHARACTERISTICS.csv"

# Load the datasets
dat_scan_df = pd.read_csv(datscan_path)
updrs_df = pd.read_csv(updrs_path)
non_motor_df = pd.read_csv(non_motor_path)
subjects_df = pd.read_csv(subjects_path)

```
