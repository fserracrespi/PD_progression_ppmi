```python
import pandas as pd
from pathlib import Path

current_dir = Path.cwd()
project_root = current_dir.parents[2]
data_dir = project_root / "DATA_PPMI/DAT_SCAN/Xing_Core_Lab_-_Quant_SBR_02Mar2026.csv"
subject_path = project_root / "DATA_PPMI/Results/SUBJECT_CHARACTERISTICS.csv"

patnos = pd.read_csv(subject_path)["PATNO"].unique().tolist()
print('Number of unique patients:', len(patnos))

EVENTS=["SC","BL", "V04", "V06", "V08", "V10", "V12"]
```

```python
data=pd.read_csv(data_dir)
data=data.loc[ data['PATNO'].isin(patnos) & data['EVENT_ID'].isin(EVENTS)&(data['DATSCAN_ANALYZED']=='Yes')]
data.drop(columns=['PROTOCOL', 'PREVIOUSLY_ACQUIRED','DATSCAN_LIGAND','DATSCAN_NOT_ANALYZED_REASON','DATSCAN_OTHER_SPECIFY','DATSCAN_ANALYZED','DATSCAN_DATE'], inplace=True)

patnos_bl = set(data.loc[data['EVENT_ID'] == 'BL', 'PATNO'])
patnos_sc = set(data.loc[data['EVENT_ID'] == 'SC', 'PATNO'])

patnos_con_bl_y_sc = patnos_bl.intersection(patnos_sc)

# Eliminar los registros SC solo para esos PATNOs
data = data[
    ~((data['PATNO'].isin(patnos_con_bl_y_sc)) & (data['EVENT_ID'] == 'SC'))
]

data['EVENT_ID']= data['EVENT_ID'].replace({'SC':'BL'})

path= project_root / "DATA_PPMI/Results/DatScan_SBR.csv"
data.to_csv(path, index=False)
```
